```
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

# Total Fruits and Vegetables Imports 1997-2019

```
arm_fv_total = pd.read_csv("https://raw.githubusercontent.com/amindazad/Armenia_Fruits_Vegetables/main/Datasets/World/Veg%20Fruits%20Total/07_08_Armenia_1999_2019.csv").sort_values(by="Year")

arm_fv_total = arm_fv_total.drop(['Period', 'Period Desc.', 'Trade Flow', 
                       '2nd Partner Code', '2nd Partner','2nd Partner ISO',
                       'Customs Proc. Code','Customs','Mode of Transport Code',
                       'Mode of Transport','Alt Qty Unit Code','Alt Qty Unit','Alt Qty',
                        'Gross weight (kg)','CIF Trade Value (US$)','FOB Trade Value (US$)',
                        'Aggregate Level','Is Leaf Code','Flag'],axis=1)

arm_fv_total = arm_fv_total.replace({'Edible vegetables and certain roots and tubers':'Vegetables',
                                     'Edible fruit and nuts; peel of citrus fruit or melons' : 'Fruits',
                                     'Vegetables and certain roots and tubers; edible':'Vegetables',
                                     'Fruit and nuts, edible; peel of citrus fruit or melons' : 'Fruits', 
                 "Edible fruit, nuts, peel of citrus fruit, melons":"Fruits"})
arm_fv_total
```

```
arm_fv_total_fig_1 = px.bar(arm_fv_total, x="Year", y="Trade Value (US$)",
                            color="Commodity",barmode='overlay',
                            template="plotly_white",text="Trade Value (US$)", 
                            title="Armenia Fruits & Vegetables Import 1997-2019 (Source: UN Comtrade 2019)")

arm_fv_total_fig_1.update_traces(texttemplate='%{text:.3s}', 
                                 textposition='outside')
#rus_fv_total_fig_1.for_each_annotation(lambda a: a.update(text=a.text.split("=")[1]))
arm_fv_total_fig_1.for_each_trace(lambda t: t.update
                                  (name=t.name.replace("Commodity=", "")))

arm_fv_total_fig_1.show()
```

# Fruits imports by item 1999-2019

```
arm_fruits_byitem_1999_2019_801 = pd.read_csv('https://raw.githubusercontent.com/amindazad/Armenia_Fruits_Vegetables/main/Datasets/World/Fruits%20Items/0801_Armenia_World_1999_2019.csv')
arm_fruits_byitem_1999_2019_802 = pd.read_csv('https://raw.githubusercontent.com/amindazad/Armenia_Fruits_Vegetables/main/Datasets/World/Fruits%20Items/0802-0803_Armenia_World_1999_2019.csv')
arm_fruits_byitem_1999_2019_808 = pd.read_csv('https://raw.githubusercontent.com/amindazad/Armenia_Fruits_Vegetables/main/Datasets/World/Fruits%20Items/0808-0812_Armenia_World_1999_2019.csv')
arm_fruits_byitem_1999_2019_813 = pd.read_csv('https://raw.githubusercontent.com/amindazad/Armenia_Fruits_Vegetables/main/Datasets/World/Fruits%20Items/0813-0814_Armenia_World_1999_2019.csv')
arm_fruits_byitem_1999_2019_804 = pd.read_csv('https://raw.githubusercontent.com/amindazad/Armenia_Fruits_Vegetables/main/Datasets/World/Fruits%20Items/0804-0807_Armenia_World_1999_2019.csv')
arm_fruits_byitem_1999_2019_80390 = pd.read_csv('https://raw.githubusercontent.com/amindazad/Armenia_Fruits_Vegetables/main/Datasets/World/Fruits%20Items/080390_Armenia_World_1999_2019.csv')

arm_fruits_byitem_1999_2019_80390 = arm_fruits_byitem_1999_2019_80390.replace({'Fruit, edible; bananas, other than plantains, fresh or dried':'Bananas, including plantains'})

arm_fruits_byitem_list = [arm_fruits_byitem_1999_2019_801,
                          arm_fruits_byitem_1999_2019_802,
                          arm_fruits_byitem_1999_2019_804,
                          arm_fruits_byitem_1999_2019_808,
                          arm_fruits_byitem_1999_2019_813,
                          arm_fruits_byitem_1999_2019_80390]

arm_fruits_byitem_1999_2019 = pd.concat(arm_fruits_byitem_list).sort_values(
    by=['Year', 'Trade Value (US$)'], ascending=False)

arm_fruits_byitem_1999_2019 = arm_fruits_byitem_1999_2019.drop(['Period', 'Period Desc.', 'Trade Flow', 
                       '2nd Partner Code', '2nd Partner','2nd Partner ISO',
                       'Customs Proc. Code','Customs','Mode of Transport Code',
                       'Mode of Transport','Alt Qty Unit Code','Alt Qty Unit','Alt Qty',
                       'Gross weight (kg)','CIF Trade Value (US$)','FOB Trade Value (US$)',
                       'Aggregate Level','Is Leaf Code','Flag'],axis=1)

arm_fruits_byitem_1999_2019['Average_Price'] = arm_fruits_byitem_1999_2019['Trade Value (US$)']/arm_fruits_byitem_1999_2019['Netweight (kg)']

arm_fruits_byitem_1999_2019.head()
```

```
arm_fruits_byitems_world_bar_overlay = px.bar(arm_fruits_byitem_1999_2019, x="Year", 
                               y="Trade Value (US$)", 
                               color='Commodity',
                               template='plotly_white',
                               title="Armenia Fruits fruits & vegetables by Category (Source: UN Comtrade 2019)",
                               barmode='overlay')

arm_fruits_byitems_world_bar_overlay.for_each_trace(lambda t: t.update
                                  (name=t.name.replace("Commodity=", "")))

arm_fruits_byitems_world_bar_overlay.show()
```
